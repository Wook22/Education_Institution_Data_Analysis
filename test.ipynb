# Data Analysis on Public Education Dataset

```python
import pandas as pd
import matplotlib as plt
```

```python
test_df = pd.read_csv("data/MERGED2021_22_PP.csv",low_memory=False)

test_df
```

```python
test_df.columns
```

```python
pd.set_option('display.max_rows', None)

null_sum = test_df.isnull().sum()  # Calculate the sum of null values for each column
column_count = len(test_df)  # Get the total number of rows in the DataFrame
null_percentage = (null_sum / column_count)  # Calculate the percentage of null values for each column
# Filter the results to include only columns with null percentage less than 30%
filtered_results = null_percentage[null_percentage < 0.3]
print(filtered_results.to_frame())
```

```python
null_percentage = test_df.isnull().sum()/len(test_df) *100

columns_above_threshold = null_percentage[null_percentage > 30].index.tolist()

print(columns_above_threshold)
```
